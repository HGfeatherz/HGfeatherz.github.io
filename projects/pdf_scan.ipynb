# PDF scanner/search

Here's a script i put together in order to help scour through a pile of unsorted pdf's to help find what I was looking for.
Its quite simple but really useful, it simply takes the key words, searchs for them, and then passes the name of that pdf to a csv 
file for reference.

```python
import os  
import PyPDF2  
import csv  
from tqdm import tqdm  
from PyPDF2 import PdfReader  
  
# keyword search
keywords = ['insurance', 'car']  
  
# where to search
path = 'C:/Users/searchpath'  
  
# save file location 
csv_file = 'C:/Users/savefilehere'  
  

with open(csv_file, 'w', newline='') as csvfile:  
    writer = csv.writer(csvfile)  
  
     
    writer.writerow(['File Name'] + keywords)  
  
   
    for root, dirs, files in tqdm(os.walk(path), desc='Searching for keywords in PDF files', unit='file'):  
        for filename in files:  
            if filename.endswith('.pdf'):  
                try:  
                     
                    with open(os.path.join(root, filename), 'rb') as pdf_file:  
                        
                        try:  
                            pdf_reader = PdfReader(pdf_file)  
                            num_pages = len(pdf_reader.pages)  
                            found_keywords = []  
  
                            for i in range(num_pages):  
                                page = pdf_reader.pages[i]  
                                text = page.extract_text()  
                                for keyword in keywords:  
                                    if keyword.lower() in text.lower():  
                                        found_keywords.append(keyword)  
  
                            
                            if found_keywords:  
                                writer.writerow([filename] + [True if keyword in found_keywords else False for keyword in keywords])  
                        except Exception:  
                             
                            pass  
                except PermissionError:  
                    
                    pass  
```
